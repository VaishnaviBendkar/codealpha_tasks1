```python
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')   # make sure name matches exactly!
print(df.head())
```

```python
print(df.shape)        # how many rows and columns
print(df.info())       # data types of each column
print(df.describe())
```

```python
print(df.isnull().sum())  # shows how many values are missing in each column
```

```python
# Fill missing Age with average
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Remove rows with missing Embarked
df = df.dropna(subset=['Embarked'])
```

```python
# How many people survived?
print(df['Survived'].value_counts())

# What was the average age of passengers?
print(df['Age'].mean())

# How many male vs female passengers?
print(df['Sex'].value_counts())
```
